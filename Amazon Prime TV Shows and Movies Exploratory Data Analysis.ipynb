## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data Wrangling
import ast

def safe_to_list(x):
    if pd.isna(x):
        return []
    if isinstance(x, list):
        return x
    if isinstance(x, str):
        try:
            parsed = ast.literal_eval(x)
            return parsed if isinstance(parsed, list) else []
        except:
            return []
    return []

# Remove old helper columns if notebook was rerun
titles = titles.drop(columns=['genres_list', 'countries_list'], errors='ignore')

# Remove duplicates BEFORE creating list columns
titles = titles.drop_duplicates()
credits = credits.drop_duplicates()

# Genre processing
titles['genres_list'] = titles['genres'].apply(safe_to_list)
genre_exploded = titles.explode('genres_list').rename(columns={'genres_list': 'genre'})
genre_exploded = genre_exploded[genre_exploded['genre'].notna()].copy()

# Country processing
titles['countries_list'] = titles['production_countries'].apply(safe_to_list)
country_exploded = titles.explode('countries_list').rename(columns={'countries_list': 'country'})
country_exploded = country_exploded[country_exploded['country'].notna()].copy()

# Merge datasets for actor analysis
cast_dir = credits.merge(titles[['id', 'title']], on='id', how='left')

print("Data wrangling completed successfully.")
````

### What all manipulations have you done and insights you found?

Several preprocessing steps were applied to make the dataset analysis-ready. Duplicate rows were removed from both datasets to avoid repeated entries affecting the results. Missing values in numerical columns were handled using median imputation, while missing values in categorical columns were filled using the mode or a fallback value such as "Unknown" where necessary.

The runtime column was further treated for outliers using the IQR method to reduce distortion in distribution-based analysis. In addition, list-like string columns such as genres and production countries were converted into actual Python lists and exploded into row-level values. This made it possible to perform accurate frequency analysis on genres and countries.

Finally, the titles and credits datasets were merged using the common title identifier so that actor and director analysis could be performed in the same workflow. These manipulations improved the reliability, readability, and usability of the data for structured EDA.

The main insight from the wrangling stage is that the dataset contains rich business-relevant information, but a meaningful analysis is only possible after cleaning and restructuring several fields properly.
